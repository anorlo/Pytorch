```
# from torch.utils.data.dataset import Dataset
from torch.utils.data import Dataset, DataLoader
import torch.utils.data as Data
from torchvision import transforms
import pandas as pd
from PIL import Image
import argparse
import nibabel as nib
import sys
import numpy as np

parser = argparse.ArgumentParser(description='Keras DenseUnet Training')
#  data folder
parser.add_argument('-data', type=str, default='./', help='test images')
parser.add_argument('-save_path', type=str, default='./')
#  other paras
parser.add_argument('-b', type=int, default=1)
parser.add_argument('-input_size', type=int, default=224)
parser.add_argument('-model_weight', type=str, default='')
# parser.add_argument('-model_weight', type=str, default='/diskb/ziyuan/LITS/Experiments/hybrid_model/weights.01-0.56.hdf5')
parser.add_argument('-input_cols', type=int, default=8)
parser.add_argument('-arch', type=str, default='')

#  data augment
parser.add_argument('-mean', type=int, default=48)
args = parser.parse_args("")

# idx = 0
# img = nib.load(args.data + 'volume-' + str(idx) + '.nii')
# print(img.dataobj[:,:,0])
# print(img.header['db_name'])  

class MyDataset(Dataset):
    def __init__(self):
      self.transform = transforms.Compose([transforms.ToTensor()])
      return
    def __len__(self):
      return 131
        
    def __getitem__(self, index):
      print('img'+ str(index))
      img = nib.load(args.data + 'volume-' + str(index) + '.nii')
      a = np.array(img.dataobj)
      return self.transform(a)

data  = MyDataset()
trainloader = DataLoader(data, batch_size=1, shuffle=False)
for batch_idx, images in enumerate(trainloader):
  print(type(images))
  print(images.size())
# dataiter = iter(trainloader)
# images = dataiter.next()
```
